```python
import requests 
import pandas as pd 
from bs4 import BeautifulSoup 
```

```python
#Checking whether scraping is allowed or not 
url = " https://www.nfl.com/standings/league/2020/REG"
page = requests.get(url)
```

```python
#Checking wheter all HTML can be extracted or not 
soup = BeautifulSoup(page.text,'lxml')
# soup 
```

```python
#Getting all the names in columns 
column_name = soup.find('tr')
# column_name

colName = [] 
for col in column_name:
    colData = col.text 
    colName.append(colData)
# colName

#Getting all the names in columns in a data frame  
df = pd.DataFrame(columns=colName)     
# df
```

```python
#Getting all the row data in a data frame 
row_name = soup.find_all('tr')[1:]
# row_name
rowName = []
for j in row_name:
    full_name = j.find_all('td')[0].find('div',class_ = 'd3-o-club-fullname').text.strip()
    rowData = j.find_all('td')[1:]
    row = [tr.text for tr in rowData]
    row.insert(0,full_name)
    lenght = len(df)
    df.loc[lenght] = row 
# rowName 
df 
```
